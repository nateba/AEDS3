```python
import pandas as pd 

tabela = pd.read_csv("C:/Users/nacli/OneDrive/Área de Trabalho/Faculdade/AEDS3-main/AEDS3-main/ManipulandoBase/anime_tratado.csv")
display(tabela)
```

```python
display(tabela.info())
```

```python
tabela['Aired'] = tabela['Aired'].str.replace(r'\s*to.*$', '', regex=True)
display(tabela)
```

```python
def try_parse_date(date_str):
    formats_to_try = ['%b %d, %Y', '%b, %Y']
    
    for fmt in formats_to_try:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            pass
    
    return None

df = pd.DataFrame(columns=['Aired'])
df['Aired'] = tabela['Aired']
df['Aired_converted'] = df['Aired'].apply(try_parse_date)

df['Aired']= df['Aired_converted']
df.drop(columns=['Aired_converted'], inplace=True)

tabela['Aired'] = df['Aired']


display(tabela)
```

```python
tabela.to_csv("anime_tratado.csv", index=False)
```
